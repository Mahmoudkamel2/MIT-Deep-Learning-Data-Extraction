```python
for job in jobs :
    
    a_tag = job.find('a' , class_="css-o171kl")
    title = a_tag.get_text()
    link = a_tag.attrs["href"]
    company = job.find('a' , class_="css-ipsyv7").get_text().replace('-', '').strip()
    location = job.find('span' , class_="css-16x61xq").get_text()
    try :
        last_update = job.find('div' , class_="css-eg55jf").get_text()
    except :                                                                  # Using 'try/except' due to differences in HTML 'class' names
        last_update = job.find('div' , class_="css-1jldrig").get_text()   

    job_type = job.find('span' , class_="css-uc9rga eoyjyou0").get_text()
    experience = [s.get_text().replace('·', '').strip() for s in job.find_all('span') if 'Yrs' in s.get_text()][0]
    

    data = {
        'Title' : title,
        'Link' : link ,
        'Company Name' : company,
        'Location' : location,
        'Posted' : last_update,
        'Type' : job_type,
        'Experience' :  experience   
    }

    print(data)
    print("_"*40)
```

```python
i = 0 
f = open("Wuzzuf_jobs.csv" ,'w' ,newline='', encoding='utf-8-sig')
writer = DictWriter(f,['Title', 'Company Name', 'Location' ,'Posted' ,'Type' ,'Experience' ,'Link'])
writer.writeheader()
while True :
    response = requests.get(f"https://wuzzuf.net/search/jobs/?q=data%20science&start= {i}")
    soup = BeautifulSoup(response.text)
    jobs = soup.find_all('div' ,class_="css-pkv5jc")

    if len(jobs) == 0:
        print("Finished Scrapping")
        break
        
    
    for job in jobs :
        
        a_tag = job.find('a' , class_="css-o171kl")
        title = a_tag.get_text()
        link = a_tag.attrs["href"]
        company = job.find('a' , class_="css-ipsyv7").get_text().replace('-', '').strip()
        location = job.find('span' , class_="css-16x61xq").get_text()
        try :
            last_update = job.find('div' , class_="css-eg55jf").get_text()
        except :                                                                  # Using 'try/except' due to differences in HTML 'class' names
            last_update = job.find('div' , class_="css-1jldrig").get_text()   
    
        job_type = job.find('span' , class_="css-uc9rga eoyjyou0").get_text()
        
        try :
            experience = [s.get_text().replace('·', '').strip() for s in job.find_all('span') if 'Yrs' in s.get_text()][0]
        except :
            experience = None 
        
        
    
        data = {
            'Title' : title,
            'Link' : link ,
            'Company Name' : company,
            'Location' : location,
            'Posted' : last_update,
            'Type' : job_type,
            'Experience' :  experience   
        }
        
        writer.writerow(data)
        
    i += 1

f.close()
```
