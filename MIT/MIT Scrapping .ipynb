```python
for vid in Videos :
    data = {}

    person = vid.find('span', class_="person-name")
    position = vid.find('span',class_='person-position')
    title = vid.find("a")

    try:
        data["title"] = title.get_text().strip()
    except:
        data["title"] = None

    try:                                                   #علشان لو الاسم مش موجود يعمله سكيب مش يعمل Erorr
        data["person"] = person.get_text()
    except:
         data["person"] = None

    try:   
        data["position"] = position.get_text().replace(',', '').strip()
    except:
        data["position"] = None
        
    try:
        video_link = vid.find("a")["href"]
        data["video_link"] = video_link
    except:
        data["video_link"] = None
    
    print(data) 
    print("_"*30)
```

```python
import csv 

with open("MIT Videos.csv", 'w', newline ='') as f :
    writer = csv.DictWriter(f, ['title','person','position','video_link'])
    writer.writeheader()
    
    for vid in Videos :
        data = {}
        person = vid.find('span', class_="person-name")
        position = vid.find('span',class_='person-position')
        title = vid.find("a")
        try:
            data["title"] = title.get_text().strip()
        except:
            data["title"] = None
        try:                                                   #علشان لو الاسم مش موجود يعمله سكيب مش يعمل Erorr
            data["person"] = person.get_text()
        except:
             data["person"] = None
        try:   
            data["position"] = position.get_text().replace(',', '').strip()
        except:
            data["position"] = None
            
        
        video_link = vid.find("a")["href"]
        data["video_link"] = video_link
        
        writer.writerow(data)
```
